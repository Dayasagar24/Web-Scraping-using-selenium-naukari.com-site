```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Noukari.com
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.naukri.com/')
```

```python
destignation = driver.find_element(By.CLASS_NAME,"suggestor-input")
destignation.send_keys("Data Analyst")
```

```python
locations = driver.find_elements(By.CLASS_NAME,"suggestor-input")
locations[1].send_keys("Mumbai")
```

```python
search = driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
title_tags = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/h2/a')
for i in title_tags:
    title=i.text
    job_title.append(title)
job_title=job_title[0:15]
print(job_title)
len(job_title)
                                  
```

```python
location_tags = driver.find_elements(By.CLASS_NAME,'locWdth')
for i in location_tags:
    title=i.text
    job_location.append(title)
job_location=job_location[0:15]
print(job_location)
len(job_location)
                   
```

```python
company_name_tags= driver.find_elements(By.XPATH,'//div/span/a[@class=" comp-name mw-25"]')
for i in company_name_tags:
    title=i.text
    company_name.append(title)
company_name=company_name[0:15]
print(company_name)
len(company_name)
                   
```

```python
experience_required_tags = driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in experience_required_tags:
    title=i.text
    experience_required.append(title)
experience_required=experience_required[0:15]
print(experience_required)
len(experience_required)
```

```python
df=pd.DataFrame({"Job Title":job_title,"Job location":job_location,"Company Name":company_name,"experience_required":experience_required})
df
```

```python
df.to_excel("Naukri_Web_Scraping.xlsx")
```
